```python
from estnltk.text import Text
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

The average read count for each word found in title. The file is made using words_in_title.py found in the python folder.

```python
avg_reads = pd.read_csv("../data/count_for_each_title.csv", sep=",", names=['word', 'reads', 'freq'], encoding="utf-8")

def method(row):
    return row.reads / row.freq

avg_reads['avg_reads'] = avg_reads.apply(lambda i: method(i), axis=1)
```

Top 10 names in articles title by average read count (50 to 100 occurences)

```python
exclude = 'Blogi|^[a-zöäüõ]|\'ja\'|[:»«]|\',\'|\'.\'*'
avg_reads[(avg_reads['freq'] >= 50) & (avg_reads['freq'] <= 100) & (~avg_reads['word'].str.contains(pat = exclude, na=False, regex = True))].sort_values(by="avg_reads", ascending=False).head(10)
```

The average read count for bigrams in title. Made by bigram_read_count.py in the python folder

```python
avg_reads = pd.read_csv("../../data/bigram_read_count.csv", sep=",", names=['bigram', 'reads', 'freq'], encoding="utf-8")

def find_avg_reads(row):
    return row.reads / row.freq

avg_reads['avg_reads'] = avg_reads.apply(lambda i: find_avg_reads(i), axis=1)
```

Top 10 bigrams by average read count. (5 to 10 occurences)

```python
exclude = '\'on\'|[A-Z]|\'ja\'|[:»«]|\',\'|\'.\''
avg_reads[(avg_reads['freq'] >= 5) & (avg_reads['freq'] < 10) & (~avg_reads['bigram'].str.contains(pat = exclude, regex = True))].sort_values(by="avg_reads", ascending=False).head(10)
```
